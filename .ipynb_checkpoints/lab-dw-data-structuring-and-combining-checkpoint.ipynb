Combine the data from the three dataframes into a single dataframe, named "customer_data", using appropriate merging, concatenating, and joining techniques.

Verify that the customer_data dataframe contains all the rows and columns from the three original dataframes.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import statsmodels.api as sm
from scipy.stats import boxcox
pd.options.display.max_rows = 100
```

```python
def load_original_data():
    file1 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv')
    file2 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2_clean.csv')
    file3 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3_clean.csv')
    return pd.concat([file1,file2,file3], axis=0)
```

```python
customer_data_df = load_original_data()
```

```python
customer_data_df.head
```

```python
def lower_case_column_names(customer_data_df):
    customer_data_df.columns = [i.lower() for i in customer_data_df.columns]
    return customer_data_df
```

```python
customer_data_df = lower_case_column_names(customer_data_df)
```

```python
customer_data_df.columns.unique()
```

```python
def rename_columns(customer_data_df):
    customer_data_df.rename(columns={'st':'states'}, inplace = True)
    customer_data_df.columns = [i.replace(' ', '_') for i in customer_data_df.columns]
    return customer_data_df
```

```python
customer_data_df = rename_columns(customer_data_df)
```

```python
customer_data_df
```

```python
customer_data_df.reset_index(inplace=True)
```
